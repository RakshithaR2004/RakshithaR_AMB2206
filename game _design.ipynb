```python
print("welcome to farming game yard")

print("press 1- sowing")
print("'press 2- irrigating'")
print("press 3- harvesting")
print("press 4- exit")


choice= int(input("enter your choice"))
crop=0
point=0

while (True):
   if choice==1:
      crop+=1
      print("sowing is done")
   elif choice==2:
      if crop >0:
        print("irrigated")
      else:
        print("no crops to irrigate")
        print("please plant crops before irrrigating")
   elif choice==3:
      if crop >0:
        print("harvested")
        point=crop*10
        crop=0
      else:
        print("no crops to harvest")
        print("please plant crops before harvesting")
   elif choice==4:
      print("game over")
      print("total score=",point)
      break
   else:
      print("invalid choice")
```

```python
from typing import Counter
import random
#initialize inventory and money
inventory={}
total_money=0
fish_prices={"catla":12,"rohu":17,"mrigal":11}

print("welcome to FISHING BUSINESS GAME")
print("press 1- FISHING")
print("press 2- SHOW INVENTARY")
print("press 3- SELL FISH")
print("press 4- END GAME")

while(True):
  choice=int(input("choose an potion"))

  if choice==1:
    fish=random.choice(list(fish_prices.keys()))
    inventory[fish]=inventory.get(fish,0)+1
    print(f"you cought a fish !")

  elif choice==2:
    if inventory=={}:
      print("no fish")

    else:
      print("your inventory: ")
      for fish, count in inventory.items():
        print(f"{fish}:{count}")

  elif choice==3:
    if inventory=={}:
      print("no fish to sell")
    else:
      earnings=0
      for fish,count in inventory.items():
        earnings+=fish_prices[fish]*count
        total_money+=earnings
        inventory.clear()
        print(f"you sold fish for${earnings}")

  elif choice==4:
    print(f"final money earned = ${total_money}")
    print("thanks for playing")
    break
  else:
    print("invalid choice.try again")



```
