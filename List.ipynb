```python
r=float(input("Enter the radius= "))
a=3.14*r*r
p=2*3.14*r
# x is to give option to user either to choose 1 or 2
# 1 for area of circle
# 2 for perimeter of circle
x=float(input("Enter a no.= "))
if(x==1):
  print("Area of circle is ",a)
elif(x==2):
  print("Perimeter of circle is ",p)
else:
  print("Enter either 1 or 2")
```
